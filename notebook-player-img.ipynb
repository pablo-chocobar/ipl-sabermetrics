```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:107.0) Gecko/20100101 Firefox/107.0"
LANGUAGE = "en-US,en;q=0.5"
```

```python
session = requests.Session()
session.headers['User-Agent'] = USER_AGENT
session.headers['Accept-Language'] = LANGUAGE
session.headers['Content-Language'] = LANGUAGE
```

```python
def team_url(team):
    url = "https://www.iplt20.com/teams/" + team + "/squad"
    return url
```

```python
team_list = ["chennai-super-kings", "delhi-capitals" ,"gujarat-titans", "kolkata-knight-riders", "lucknow-super-giants", "mumbai-indians", 
            "punjab-kings", "rajasthan-royals", "royal-challengers-bangalore" ,"sunrisers-hyderabad"]
```

```python
import os
import shutil
```

```python
parent_dir = r"D:\Projects\000000000000000000000000000000\analysis\images"
```

```python
for i in team_list:
    path = os.path.join(parent_dir, i)
    os.mkdir(path)
    print("created: ", path)
    os.chdir(path)
    html = session.get(team_url(i))
    
    links = []
    names = []
    
    soup = bs(html.text, "html.parser")
    p = soup.findAll("div", attrs = {"class": "ih-p-img"})
    
    for i in range(len(p)):
        links.append(p[i].find("img").attrs["src"])
        names.append(p[i].find("h2").get_text())
        
    for i in range(len(links)):
        imgurl = links[i]
        filename = names[i]+".png"

        r = requests.get(imgurl, stream = True)
        if r.status_code == 200:

            r.raw.decode_content = True

            with open(filename,'wb') as f:
                shutil.copyfileobj(r.raw, f)

            print('Image sucessfully Downloaded: ',filename)
        else:
            print('Image Couldn\'t be retreived')
```
